```python
import matplotlib.pyplot as plt

# Input voltages
input_voltages = [14.0, 14.5, 15.0, 15.5, 16.0, 16.5, 17.0, 17.5, 18.0]

# Output voltages at 6V and 12V
output_voltage_6V = [5.963, 5.977, 5.988, 6.002, 6.024, 6.031, 6.042, 6.055, 6.067]
output_voltage_12V = [11.797, 11.848, 11.896, 11.931, 11.974, 11.996, 12.016, 12.058, 12.093]

# Create the plot
plt.plot(input_voltages, output_voltage_6V, label='Output Voltage (6V)')
plt.plot(input_voltages, output_voltage_12V, label='Output Voltage (12V)')

# Add labels and title
plt.xlabel('Input Voltage (V)')
plt.ylabel('Output Voltage (V)')
plt.title('Line Variation of Linear Power Supply')

# Add a legend
plt.legend()

# Display the plot
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
load_resistor = [5, 6, 10, 47, 100, 330, 470, 560, 810, 2200, 3900, 5600, 10000]
output_voltage_6V = [5.80, 5.90, 5.945, 5.993, 5.999, 6.002, 6.002, 6.008, 6.008, 6.008, 6.009, 6.009, 6.009]
output_voltage_12V = [11.58, 11.73, 11.921, 11.937, 11.959, 11.977, 11.981, 11.984, 11.986, 11.990, 11.992, 11.995, 11.996]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(load_resistor, output_voltage_6V, label='Output Voltage (6V)')
plt.plot(load_resistor, output_voltage_12V, label='Output Voltage (12V)')

# Add labels and title
plt.xscale('log')  # Set the x-axis to logarithmic scale for better visualization
plt.xlabel('Load Resistor (ohms)')
plt.ylabel('Output Voltage (V)')
plt.title('Load Variation of Linear Power Supply')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
